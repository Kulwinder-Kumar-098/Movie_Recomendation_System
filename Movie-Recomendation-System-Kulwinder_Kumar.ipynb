```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
```

```python
def convert3(obj):
    L=[]
    counter =0 
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter +=1
        else :
            break   
    return L 
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']== 'Director':
            L.append(i['name'])
            break
    return L 
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
def recommend(movie):
    try:
        movie_index = new_df[new_df['title'] == movie].index[0]
        distances = similarity[movie_index]
        movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

        for i in movies_list:
            print(new_df.iloc[i[0]].title)
    except IndexError:
        print(f"Movie '{movie}' not found in the dataset.")
```

```python
from difflib import get_close_matches

def recommended(movie):
    movie_titles = new_df['title'].tolist()
    close_matches = get_close_matches(movie, movie_titles, n=1, cutoff=0.5)
    
    if close_matches:
        movie_match = close_matches[0]
        movie_index = new_df[new_df['title'] == movie_match].index[0]
        distances = similarity[movie_index]
        movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

        print(f"Recommendations for '{movie_match}':")
        for i in movies_list:
            print(new_df.iloc[i[0]].title)
    else:
        print(f"No close match found for '{movie}'. Please try another title.")

```
